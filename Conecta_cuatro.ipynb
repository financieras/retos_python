```python
import random

def coin_flip():
    return random.choice(['cara', 'cruz'])

def play_game(num_rounds):
    scores = [0] * num_players

    for round in range(num_rounds):
        print(f"Ronda {round + 1}:")
        for player in range(num_players):
            result = coin_flip()
            if result == 'cara':
                scores[player] += 1
            print(f"Jugador {player + 1}: {result}")

    print("\nResultados finales:")
    for player in range(num_players):
        print(f"Jugador {player + 1}: {scores[player]} puntos")

    winners = [str(i + 1) for i, score in enumerate(scores) if score == max(scores)]
    if len(winners) == 1:
        print(f"\nEl jugador {', '.join(winners)} gana!")
    else:
        print(f"\nHay un empate entre los jugadores {', '.join(winners)}!")

# Configuración del juego
num_players = int(input("Ingrese el número de jugadores: "))
num_rounds = int(input("Ingrese el número de rondas: "))

# Jugar el juego
play_game(num_rounds)
```

```python
import numpy as np

# Crear el tablero
board = np.full((6, 7), "·", dtype=str)

# Función para colocar una ficha en el tablero
def drop_piece(board, row, col, piece):
    board[row][col] = piece

# Función para obtener la posición más baja disponible en una columna
def get_next_row(board, col):
    for row in range(5, -1, -1):
        if board[row][col] == "·":
            return row
    return -1

# Función para verificar si hay un ganador
def check_winner(board, piece):
    # Verificar en horizontal
    for row in range(6):
        for col in range(4):
            if board[row][col] == piece and board[row][col+1] == piece and board[row][col+2] == piece and board[row][col+3] == piece:
                return True

    # Verificar en vertical
    for row in range(3):
        for col in range(7):
            if board[row][col] == piece and board[row+1][col] == piece and board[row+2][col] == piece and board[row+3][col] == piece:
                return True

    # Verificar en diagonal (izquierda a derecha)
    for row in range(3):
        for col in range(4):
            if board[row][col] == piece and board[row+1][col+1] == piece and board[row+2][col+2] == piece and board[row+3][col+3] == piece:
                return True

    # Verificar en diagonal (derecha a izquierda)
    for row in range(3):
        for col in range(3, 7):
            if board[row][col] == piece and board[row+1][col-1] == piece and board[row+2][col-2] == piece and board[row+3][col-3] == piece:
                return True

    return False

# Función para verificar si el tablero está lleno
def is_board_full(board):
    return np.all(board != "·")

# Función para imprimir el tablero
def print_board(board):
    print(np.flip(board, 0))

# Juego automático
current_player = "X"

while True:
    if current_player == "X":
        # Jugada inteligente del jugador X
        available_cols = np.where(board[0] == "·")[0]
        for col in available_cols:
            temp_board = np.copy(board)
            row = get_next_row(temp_board, col)
            drop_piece(temp_board, row, col, current_player)
            if check_winner(temp_board, current_player):
                drop_piece(board, row, col, current_player)
                break
        else:
            # Si no hay una jugada ganadora, realizar una jugada aleatoria
            col = np.random.choice(available_cols)
            row = get_next_row(board, col)
            drop_piece(board, row, col, current_player)
    else:
        # Jugada aleatoria del jugador O
        col = np.random.choice(np.where(board[0] == "·")[0])
        row = get_next_row(board, col)
        drop_piece(board, row, col, current_player)

    # Imprimir el tablero
    #print_board(board)
    for fila in board:
        print(*fila)
    print()

    # Verificar si hay un ganador
    if check_winner(board, current_player):
        print(f"¡El jugador {current_player} gana!")
        break

    # Verificar si el tablero está lleno
    if is_board_full(board):
        print("¡Empate!")
        break

    # Cambiar al siguiente jugador
    current_player = "O" if current_player == "X" else "X"
```
